```python
import pandas as pd
df=pd.DataFrame(iris.data,columns=iris.feature_names)
df["target"]=iris.target
df.head()
```

```python
import matplotlib.pyplot as plt

df1=df[df["target"]==0]
df2=df[df["target"]==1]
df3=df[df["target"]==2]

plt.scatter(df1[["sepal length (cm)"]],df1[["sepal width (cm)"]])
plt.scatter(df2[["sepal length (cm)"]],df2[["sepal width (cm)"]])
plt.scatter(df3[["sepal length (cm)"]],df3[["sepal width (cm)"]])
```

```python
df1=df[df["target"]==0]
df2=df[df["target"]==1]
df3=df[df["target"]==2]

plt.title("Setosa VS Versicolor (Sepal)")
plt.scatter(df1[["sepal length (cm)"]],df1[["sepal width (cm)"]])
plt.scatter(df2[["sepal length (cm)"]],df2[["sepal width (cm)"]])
```

```python
df1=df[df["target"]==0]
df2=df[df["target"]==1]
df3=df[df["target"]==2]

plt.title("Setosa VS Versicolor (Petal)")
plt.scatter(df1[["petal length (cm)"]],df1[["petal width (cm)"]])
plt.scatter(df2[["petal length (cm)"]],df2[["petal width (cm)"]])
```

```python
from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure(1, figsize=(8, 6))
ax = Axes3D(fig, elev=-100, azim=120)

df1=df[df["target"]==0]
df2=df[df["target"]==1]
df3=df[df["target"]==2]

ax.scatter(df1[["petal length (cm)"]],df1[["petal width (cm)"]])
ax.scatter(df2[["petal length (cm)"]],df2[["petal width (cm)"]])
ax.scatter(df3[["petal length (cm)"]],df3[["petal width (cm)"]])
```

```python
from sklearn.svm import SVC
model=SVC(gamma=10)
```

```python
model.fit(df[["sepal length (cm)","sepal width (cm)","petal length (cm)","petal width (cm)"]],df.target)
```

```python
model.predict([[4.1,3.5,1.3,0.2]])
```

```python
model.score(df[["sepal length (cm)","sepal width (cm)","petal length (cm)","petal width (cm)"]],df.target)
```
