```python
import pandas as pd
```

```python
df=pd.read_csv("mob.csv")
df
```

```python
from sklearn.preprocessing  import LabelEncoder
```

```python
le=LabelEncoder()
```

```python
df["Processor_n"]=le.fit_transform(df["Processor"])
df["Brand_n"]=le.fit_transform(df["Brand"])
df
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model=LinearRegression()
```

```python
model.fit(df[["RAM","Processor_n","Brand_n"]],df.Price)
```

```python
model.predict([[7,2,0]])
```
