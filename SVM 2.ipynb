```python
import pandas as pd
df=pd.DataFrame(iris.data,columns=iris.feature_names)
df["target"]=iris.target
df
```

```python
df1=df[0:50]
df2=df[50:100]
df3=df[100:150]
```

```python
import matplotlib.pyplot as plt
plt.title("Sepal Data")
plt.scatter(df1["sepal length (cm)"],df1["sepal width (cm)"],color="red")
plt.scatter(df2["sepal length (cm)"],df2["sepal width (cm)"],color="green")
plt.scatter(df3["sepal length (cm)"],df3["sepal width (cm)"],color="blue")

```

```python
import matplotlib.pyplot as plt
plt.title("Sepal Data")
plt.scatter(df1["petal length (cm)"],df1["petal width (cm)"],color="red")
plt.scatter(df2["petal length (cm)"],df2["petal width (cm)"],color="green")
plt.scatter(df3["petal length (cm)"],df3["petal width (cm)"],color="blue")
```

```python
from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure(1, figsize=(8, 6))
ax = Axes3D(fig, elev=-120, azim=100)


ax.scatter(df1[["petal length (cm)"]],df1[["petal width (cm)"]])
ax.scatter(df2[["petal length (cm)"]],df2[["petal width (cm)"]])
ax.scatter(df3[["petal length (cm)"]],df3[["petal width (cm)"]])
```
